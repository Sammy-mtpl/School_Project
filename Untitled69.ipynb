```python
# Main Code

import csv
from os.path import isfile as check_file
from subprocess import run as terminal




developer = True



def error_handler(e):
    print("Some Unexpected error occured...")
    
    if developer:
        print("Developer Mode: On")
        print(e)
        
    else:
        print("Developer Mode: Off")
 

def clear():
    terminal("cls", shell=True)
        
        
def addingBooks():
    name = input("Enter Name of Book: ")
    genre = input("Enter Genre(Optional): ")
    quantity = int(input("Enter Quantity: "))
    
    if name and quantity > 0:
        add_book(name, genre, quantity)
        
    else:
        print("You must enter name of book and quantity should be ")

        




        
        
def add_book(name, genera, quantity=1):
    if quantity > 0:
        try:
            # Syntax of adding in main.csv
            #
            # [Name(in lowerCase with no spaces), Genera(in lowerCase with "_" replaced with spaces), Quantity (int), No. of Books Issued(int)]
            
            main_database_ki_csv = open(f"main.csv", "a", newline="")
            
            writer = csv.writer(main_database_ki_csv)
            
            writer.writerow([name.lower().replace(" ", ""), genera, quantity, 0]) # 4th value is 0 because a book can't be issued the moment it is added.
            
            main_database_ki_csv.close()
            
            return 1

        
        
        except Exception as e:
            error_handler(e)
            return 0

    else:
        print("Make sure that quantity is alteast 1")
        return 0

    
    
        
def searchingbook():
    print("Enter either Name or Genre of book")
    name = input("Name: ")
    genre = input("Genre: ")
    print("\n\n")
    
    if name != "" or genre != "":
        if name == "":
            search_books(book_name=None, genre=genre)
            
        elif genre == "":
            search_books(book_name=name, genre=None)
            
        else:
            search_books(book_name=name, genre=genre)
            
    else:
        print("Both Name and Genre are not entered...")
    


def search_books(book_name=None, genre=None):
    # Open and read the CSV file
    try:
        books = []
        with open('main.csv', newline='', encoding='utf-8') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                books.append(row)

        # Filter books by book_name and genre
        results = []

        for book in books:
            # Check if book_name is provided and matches
            # print(book)
            if book_name:
                book_name = book_name.replace(" ", "")
                if book_name.lower() != book["bookName"].lower():
                    continue  # Skip this book if it doesn't match the name

            # Check if genre is provided and matches
            if genre:
                if genre.lower() != book["genre"].lower():
                    continue  # Skip this book if it doesn't match the genre

            results.append(book)
    
        for i in results:
            print("Name: ",i["bookName"])
            print("Genre: ",i["genre"])
            print("Total: ",i["quantity"])
            print("Issued: ",i["issued"])
            print("\n")
    
    except Exception as e:
            error_handler(e)
            return 0
        
        
def showAllBooks():
    with open('main.csv', newline='', encoding='utf-8') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                print("Name: ",row["bookName"])
                print("Genre: ",row["genre"])
                print("Total: ",row["quantity"])
                print("Issued: ",row["issued"])
                print("\n")
                
                
def delete_row_from_csv(column_name='bookName'):
    value_to_delete = input("Enter name of book you want to delete: ")
    
    surity = input("Are you sure you want delete the specified record? [Y/N]: ")
    if surity == "Y":
        # Read all rows from the CSV file
        rows = []
        records_found = False
        with open('main.csv', 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                # Keep rows that do not contain the value in the specified column
                if row[column_name] != value_to_delete.lower().replace(" ", ""):
                    rows.append(row)
                else:
                    records_found = True
                

        # Write the filtered rows back to the CSV file
        with open('main.csv', 'w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=rows[0].keys())
            writer.writeheader()
            writer.writerows(rows)
            
        if records_found:
            print("Value deleted successfully")
        else:
            print("No Such Record Found....")
    else:
        print("File not deleted....")
        
    
################################################ Menu Making ########################################################
    

def clear():
    terminal("cls", shell=True)
    

def menu():
    print("\n\nSchool Records")
    print("Enter command")
    admin()
        
  

commands = {1: addingBooks, 2: searchingbook, 3:showAllBooks, 4:delete_row_from_csv}
        
        
def admin():
    clear()
    while True:
        try:
            print("""
    [1] Create a Book Record
    [2] Search for a Book
    [3] View all Books available
    [4] Delete Book
    [5] Exit
    """)
            u_inp = int(input("\nEnter your command: "))

            if u_inp==5:
                break

            commands[u_inp]()


        except Exception as e:
            if e == IndexError:
                print("This Function is yet to be developed")
                
            else:
                print("Incorrect command.")
                print("Try Again.")
                print(e)
                
                
menu()
```
